```python
# Twitter credentials for project
consumer_key = ""
consumer_secret = ""
access_token= ""
access_token_secret= ""
```

```python
# Twitter authentication and the connection to Twitter Streaming API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
# create list of words to search for in tweets
#https://docs.tweepy.org/en/stable/api.html
words = ['vegan produce', 'vegan', 'vegetarian', 'plant based', '@vegan', '@farm', '@farmer', 'meat free', 'dairy free', 'vegan farming' , 'milk alternative', 'meat alternative', 'meat replacement', 'vegan friendly']
train = pd.DataFrame()
```

```python
# search only last 7 days of tweets
# https://developer.twitter.com/en/docs/twitter-api/v1/tweets/search/api-reference/get-search-tweets

for word in words:
    tweets = tweepy.Cursor(api.search_tweets, q=word+'-filter:retweets', lang='en', tweet_mode='extended').items(180)
    for tweet in tweets:
        temp = pd.DataFrame.from_dict(tweet._json, orient='index')
        temp = temp.T
        temp = temp[['created_at', 'id', 'full_text','lang']]
        temp.loc[:,'user'] = tweet._json['user']['screen_name']
        temp['User Location'] = tweet._json['user']['location']
        train= pd.concat([train, temp], axis=0)
train.reset_index(inplace=True, drop=True)
```

## Basic Feature Extraction

Extract features from tweet data to identify number of elements contained in tweets. 

```python
#create a new feature for wordcount
tweets_df["word_count"] = tweets_df["tweet"].apply(lambda x : len(str(x).split()))

# create a new feature for tweet character legnth
tweets_df["char_count"] = tweets_df["tweet"].str.len()

#create stopwords
stop = stopwords.words("english")

# if a stopword is present, save into this list
tweets_df["stopwords"] = tweets_df["tweet"].apply(lambda x : len([x for x in x.split() if x in stop]))

# number of hashtags
tweets_df['hashtag_no'] = tweets_df['tweet'].apply(lambda x : len([x for x in x.split() if x.startswith("#")]))

#extract hashtags
tweets_df['hashtag'] = tweets_df['tweet'].apply(lambda x: re.match(r'#\w+', x))

# @ names
tweets_df["at_names"] = tweets_df["tweet"].apply(lambda x : len([x for x in x.split() if x.startswith("@")]))
```

## Basic Pre-processing

https://www.analyticsvidhya.com/blog/2018/02/the-different-methods-deal-text-data-predictive-python/

This step includeds cleaning the data in order to obtain better text for analysis. This will include Removing as much data as possible that will not add anything to the analysis in order to reduce the traiing size of data. 

Steps taken:

    Ensuring text is in all lower case to avoid having multiple occuerances of the same word win different cases. removing special characters
    Removing all user ames from tweets using regular expression re.sub. this step must be done before special characters are removed as otherwise @names wont be identifiable
    Remove punctuation as this wont be used for analysis
    Remove stop words as these are generalised and do not add anything extra to the text
    Remove common and rare words that do not add anything extra to the text
    Fix abrivestiong such as 'ur' for 'your' and 'u' for 'you. 
    Remove suffices, like “ing”, “ly”, “s”, etc using Lemmatization(reverts to root word) as opposed to stemming(simply removes endings).
      

```python
#create clean tweets feature & put in lower case
tweets_df['clean_tweet'] = tweets_df['tweet'].apply(lambda x : " ".join(x.lower() for x in x.split()))
```

```python
#remove usernames form tweets before punctuation 
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].apply(lambda x: re.sub('@[^\s]+','', x))

#remove #'s' from tweets before punctuation 
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].apply(lambda x: re.sub('#[^\s]+','', x))

#removes any links from tweets
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].apply(lambda x: re.sub('http[^\s]+','', x))

#removes special words from tweets
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("레드벨벳", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("조이", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("𝕍𝕒𝕦𝕝𝕥", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("𝕎𝕒𝕘𝕠𝕟", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("órale", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("óralestreetfood", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("เนอสตว", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("แบคเทอรโอเฟจ", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("ไก", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("ไกเนอ", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("キンプる", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("박수영", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("아떼", "")
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("𐐪1003𐑂", "")


tweets_df.head()
```

```python
# remove punctuation
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].str.replace("[^\w\s]", "")
```

```python
# remove any stopwords
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].apply(lambda x : " ".join(x for x in x.split() if x not in stop))
```

Lemmatization is a more effective option than stemming because it converts the word into its 
root word, rather than just stripping the suffices. It makes use of the vocabulary and does a 
morphological analysis to obtain the root word. Therefore, we usually prefer using 
lemmatization over stemming.

```python
tweets_df['clean_tweet'] = tweets_df['clean_tweet'].apply(lambda x: " ".join([Word(word).lemmatize() for word in x.split()])) 
tweets_df['clean_tweet'].head() 
```

## Sentiment Analysis

https://www.analyticsvidhya.com/blog/2018/02/natural-language-processing-for-beginners-using-textblob/

The sentiment function of textblob returns two properties, [0]sentiment, and [1]subjectivity.

Sentiment is float which productes a result between [-1,1] where 1 means positive, -1 means negative, and 0 is neutral.  Subjectivity is also a float which lies in the range of [0,1]. Subjectivity[1] refers to personal opinion or emotion whereas objectivity[0] refers to factual information.  The higher the subjectivity, the more likely it is to be personal opinion than factual information.


```python
#identify tweet topic 
tweets_df['sentiment'] = tweets_df['clean_tweet'].apply(lambda x : TextBlob(x).sentiment[0])

#apply sentiment to tweet
tweets_df['subjectivity'] = tweets_df['clean_tweet'].apply(lambda x : TextBlob(x).sentiment[1])

tweets_df.head()
```
